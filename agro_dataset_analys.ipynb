```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv(r"data\argo_dataset\argo_dataset.csv")
df.head()
x = np.arange(len(df))
```

```python
weather_parameters = ["SOLAR_RADIATION",
                        "PRECIPITATION",
                        "WIND_SPEED",
                        "LEAF_WETNESS",
                        "HC_AIR_TEMPERATURE",
                        "HC_RELATIVE_HUMIDITY",
                        "DEW_POINT"]

n_cols = 2
n_rows = math.ceil(len(weather_parameters) / n_cols)
fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 12))
axes = axes.flatten()
for ax in axes:
    ax.axis('off')

for ax, parameter_name in zip(axes, weather_parameters):
    ax.plot(x, df[parameter_name])
    ax.axis('on')
    ax.set_xlabel("hour")
    ax.set_ylabel(parameter_name)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 12))
axes = axes.flatten()
for ax in axes:
    ax.axis('off')

x_size = 500
x_offset = int(len(df) / 2 - x_size / 2)

scaled_x = np.arange(x_offset, x_offset + x_size)
scaled_df = df.iloc[x_offset:x_offset + x_size]

for ax, parameter_name in zip(axes, weather_parameters):
    ax.plot(scaled_x, scaled_df[parameter_name])
    ax.axis('on')
    ax.set_xlabel("hour")
    ax.set_ylabel(parameter_name)

plt.tight_layout()
plt.show()
```
